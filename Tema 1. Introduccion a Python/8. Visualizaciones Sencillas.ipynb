# Frameworks sencillos para visualizar datos

Dado que estamos empezando por lo más básico, vamos a trabajar ciertos frameworks que permiten hacer visualizaciones de datos y que son más fáciles que plotly que va a ser nuestro framework de trabajo general. En este caso vamos a ver dos tipos: Matplotlib y Seaborn.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Visualizaciones usando Matplotlib

La primera librería que vamos a utilizar va a ser Matplotlib (enlace a la documentación: https://matplotlib.org/stable/contents.html), esta es una de las librerías más completas y simples que tenemos para visualización.

Aquí un enlace a las opciones que dispone Matplotlib: https://matplotlib.org/stable/gallery/index.html

### Gráfico de lineas

```python
a = np.linspace(-3,3,50)
b = np.sin(a)
 
plt.plot(a, b) 
 
plt.show()
```

```python
# Cambiar color, grosor y tipo de linea

# Trazados comunes de tipo de linea
#'-' linea sólida
#'-.'linea con puntos y rayas
#'--'línea a rayas
#':' línea punteada

plt.plot(a, b, color = 'r', linewidth = 2.5, linestyle = ':')  
plt.show()
```

```python
# Agregar más lineas al mismo grafico

c = np.cos(a)
plt.plot(a, b, color = 'r', linewidth = 2.5, linestyle = ':')
plt.plot(a, c, color = 'green', linewidth = 6, linestyle = '--')
plt.show()
```

```python
# Incluir nombres a los ejes, título y leyendas

# Cambiar forma de la figura
plt.figure(figsize = (16,9))

plt.plot(a, b, color = 'r', linewidth = 2.5, linestyle = ':')
plt.plot(a, c, color = 'green', linewidth = 6, linestyle = '--')

# Añadir nombre al eje x
plt.xlabel("Eje de abcisas", fontsize = 14)

# Añadir nombre al eje y
plt.ylabel("Eje de ordenadas", fontsize = 16, color = 'blue')

# Añadir texto en un punto concreto de la gráfica
plt.text(np.sqrt(2)/2-0.1, np.sqrt(2)/2-0.15, "Punto de corte", fontsize = 8)

# Agregar título al gráfico
plt.title("Representación de funciones", fontsize = 24)

# Agregar leyenda
plt.legend(('Función Seno', 'Función Coseno'), loc = 'upper right')

# Agregar grid
plt.grid(True)
```

Enlace a esta funcion: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html?highlight=plot#matplotlib.pyplot.plot

### Gráfico de puntos

```python
x1 = np.random.normal(loc=2, scale=1.0, size = 100)
y1 = np.random.normal(loc=4, scale=1.0, size = 100)

x2 = np.random.normal(loc=4, scale=1.0, size = 100)
y2 = np.random.normal(loc=1, scale=1.0, size = 100)

plt.figure(figsize = (16,9))

plt.scatter(x1,y1, c = "indigo", marker = '.', s = 100 )
plt.scatter(x2,y2, c = "darkorange", marker = '*', s = 20)

plt.xlabel("Eje de abcisas")
plt.ylabel("Eje de ordenadas")

plt.title("Dos muestras normales bivariantes")

plt.grid(True)
plt.show()
```

Enlace de esta funcion: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.scatter.html?highlight=scatter#

### Histogramas

```python
from matplotlib import colors

plt.figure(figsize = (12,7))

# Output de la función plt.hist
N, bins, patches = plt.hist(x1, bins = 20)

# La idea es colorear cuanto más alto más claro
fracs = N / N.max()

# Normaliza al intervalo 0-1
norm = colors.Normalize(fracs.min(), fracs.max())

# Colorea cada barra según su altura
for thisfrac, thispatch in zip(fracs, patches):
    color = plt.cm.viridis(norm(thisfrac))
    thispatch.set_facecolor(color)
    
plt.ylabel("Repeticiones dentro de cada barra")
plt.title("Histograma de valores coloreado según la altura")
    
plt.show()
```

```python
# Grafico en horizontal 
plt.figure(figsize = (12,7))

# Output de la función plt.hist
N, bins, patches = plt.hist(x1, bins = 20, orientation = "horizontal")

# La idea es colorear cuanto más alto más claro
fracs = N / N.max()

# Normaliza al intervalo 0-1
norm = colors.Normalize(fracs.min(), fracs.max())

# Colorea cada barra según su altura
for thisfrac, thispatch in zip(fracs, patches):
    color = plt.cm.viridis(norm(thisfrac))
    thispatch.set_facecolor(color)
    
plt.xlabel("Repeticiones dentro de cada barra")
plt.title("Histograma de valores coloreado según la altura")
    
plt.show()
```

Enlace de la función https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.hist.html

### Gráfico de barras

```python
cat = ['A','B','C','D']
count = [4,5,11,2]

plt.figure(figsize= (14,8))

plt.bar(cat,count,color = "mediumseagreen",width = 0.3, align = "center")

plt.xlabel("Categorias")
plt.ylabel("Repeticiones")

plt.title("Gráfico de barras")

plt.show()
```

Documentación de la función: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.bar.html

```python
# Orientar las barras de manera horizontal

plt.figure(figsize= (14,8))

# gca indica Get Current Axis y permite acceder a los ejes para modificarlos como queramos
plt.gca().invert_yaxis()

plt.barh(cat,count,color = "mediumseagreen", height = 0.3)

plt.ylabel("Categorias")
plt.xlabel("Repeticiones")

plt.title("Gráfico de barras")

plt.show()
```

```python
# Apilar otra muestra sobre la anterior
count2 = [2,3,3,5]

plt.figure(figsize= (14,8))

plt.bar(cat,count,color = "mediumseagreen",width = 0.3, align = "center",label = "Muestra1")
plt.bar(cat,count2, color = "darkorange", width = 0.3, align = "center", bottom = count, label = "Muestra2")

plt.xlabel("Categorias")
plt.ylabel("Repeticiones")

plt.title("Gráfico de barras apiladas")

plt.legend()

plt.show()
```

Documentacion de la función: https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.barh.html

### Boxplots

```python
plt.figure(figsize = (14,8))

c1 = 'steelblue'
c2 = 'darkorange'
c3 = 'mediumseagreen'
c4 = 'black'

bplot = plt.boxplot(
    x1,
    patch_artist = True,
    boxprops=dict(color=c1),
    capprops=dict(color=c2,linewidth = 2),
    whiskerprops=dict(color=c3,linewidth = 2),
    flierprops=dict(color=c1, markeredgecolor=c1),
    medianprops=dict(color=c4, linewidth = 2),
)
    
plt.title("Gráfico de Boxplot")

plt.xlabel("Muestra de una normal (2,1)")
plt.ylabel("Distribución")

# Cambiar el nombre de los ticks en los ejes
plt.xticks(ticks = [1] ,labels = [''])

plt.show()
```

```python
plt.figure(figsize = (14,8))

c1 = 'steelblue'
c2 = 'darkorange'
c3 = 'mediumseagreen'
c4 = 'black'

plt.boxplot(
    np.array([x1,x2]).T,
    positions = [1,2],
    patch_artist = True,
    boxprops=dict(color=c1),
    capprops=dict(color=c2,linewidth = 2),
    whiskerprops=dict(color=c3,linewidth = 2),
    flierprops=dict(color=c1, markeredgecolor=c1),
    medianprops=dict(color=c4, linewidth = 2),
)
    
plt.title("Gráfico de Boxplot")

plt.xlabel("Muestra de dos distribuciones normales diferentes")
plt.ylabel("Distribución")

# Cambiar el nombre de los ticks en los ejes
plt.xticks(ticks = [1,2] ,labels = ['',''])

plt.show()
```

### Pie Charts

```python
proporciones = [10,15,25,35,15]
grupos = ['G1','G2','G3','G4','G5']
sep = [0, 0, 0, 0.1 , 0]

plt.figure(figsize = (12,7))
plt.pie(proporciones, 
        labels = grupos, 
        colors = ["gold","firebrick","hotpink","indigo","lightblue"],
        explode = sep,
        shadow = True)

plt.title("Distribución de los grupos")
plt.show()
```

Enlace a la documentación: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.pie.html

### Crear múltiples imágenes en una sola imagen

```python
plt.figure(figsize = (16,16))

# Primera figura (arriba izquierda)
plt.subplot(2,2,1) # Numero de filas, Numero de columnas, Numero de figura
plt.scatter(x1,y1, c = "indigo", marker = '.', s = 100 )
plt.scatter(x2,y2, c = "darkorange", marker = '*', s = 20)

plt.xlabel("Eje de abcisas")
plt.ylabel("Eje de ordenadas")

plt.title("Dos muestras normales bivariantes")

plt.grid(True)

#Segunda figura (arriba derecha)
plt.subplot(2,2,2)

plt.bar(cat,count,color = "mediumseagreen",width = 0.3, align = "center",label = "Muestra1")
plt.bar(cat,count2, color = "darkorange", width = 0.3, align = "center", bottom = count, label = "Muestra2")

plt.xlabel("Categorias")
plt.ylabel("Repeticiones")

plt.title("Gráfico de barras apiladas")

plt.legend()

#Tercera figura (abajo izquierda)
plt.subplot(2,2,3)

bplot = plt.boxplot(
    x1,
    patch_artist = True,
    boxprops=dict(color=c1),
    capprops=dict(color=c2,linewidth = 2),
    whiskerprops=dict(color=c3,linewidth = 2),
    flierprops=dict(color=c1, markeredgecolor=c1),
    medianprops=dict(color=c4, linewidth = 2),
)
    
plt.title("Gráfico de Boxplot")

plt.xlabel("Muestra de una normal (2,1)")
plt.ylabel("Distribución")

plt.xticks(ticks = [1] ,labels = [''])

# Cuarta figura (abajo derecha)
plt.subplot(2,2,4)

plt.pie(proporciones, 
        labels = grupos, 
        colors = ["gold","firebrick","hotpink","indigo","lightblue"],
        explode = sep,
        shadow = True)

plt.title("Distribución de los grupos")
plt.show()
```

```python
# Equivalentemente con otra sintaxis

fig, axs = plt.subplots(nrows = 2, ncols = 2)

fig.set_figheight(16)
fig.set_figwidth(16)

# Primera figura (arriba izquierda)
axs[0,0].scatter(x1,y1, c = "indigo", marker = '.', s = 100 )
axs[0,0].scatter(x2,y2, c = "darkorange", marker = '*', s = 20)

axs[0,0].set_xlabel("Eje de abcisas")
axs[0,0].set_ylabel("Eje de ordenadas")

axs[0,0].set_title("Dos muestras normales bivariantes")

axs[0,0].grid(True)

#Segunda figura (arriba derecha)
axs[0,1].bar(cat,count,color = "mediumseagreen",width = 0.3, align = "center",label = "Muestra1")
axs[0,1].bar(cat,count2, color = "darkorange", width = 0.3, align = "center", bottom = count, label = "Muestra2")

axs[0,1].set_xlabel("Categorias")
axs[0,1].set_ylabel("Repeticiones")

axs[0,1].set_title("Gráfico de barras apiladas")

axs[0,1].legend()

#Tercera figura (abajo izquierda)
axs[1,0].boxplot(
    x1,
    patch_artist = True,
    boxprops=dict(color=c1),
    capprops=dict(color=c2,linewidth = 2),
    whiskerprops=dict(color=c3,linewidth = 2),
    flierprops=dict(color=c1, markeredgecolor=c1),
    medianprops=dict(color=c4, linewidth = 2),
)
    
axs[1,0].set_title("Gráfico de Boxplot")

axs[1,0].set_xlabel("Muestra de una normal (2,1)")
axs[1,0].set_ylabel("Distribución")

plt.sca(axs[1, 0])
plt.xticks(ticks = [1] ,labels = [''])

# Cuarta figura (abajo derecha)
axs[1,1].pie(proporciones, 
             labels = grupos, 
             colors = ["gold","firebrick","hotpink","indigo","lightblue"],
             explode = sep,
             shadow = True)

axs[1,1].set_title("Distribución de los grupos")

fig.show()
```

## Visualizaciones usando Seaborn

Seaborn es una librería construida sobre matplotlib que facilita a través de funciones propias la creación de gráficos e incluye alternativas para análisis estadístico.

Enlace a la librería: https://seaborn.pydata.org/

Enlace a las paletas de colores de seaborn : https://seaborn.pydata.org/tutorial/color_palettes.html

Enlace a los estilos de seaborn: https://seaborn.pydata.org/generated/seaborn.set_style.html

### Gráfico de lineas

```python
flights = sns.load_dataset("flights")
flights
```

```python
sns.set_style("whitegrid")
```

```python
grouped = pd.DataFrame(flights.groupby("year")["passengers"].mean())
plt.figure(figsize = (14,8))
sns.lineplot(data = grouped, x = "year", y = "passengers")
plt.show()
```

```python
plt.figure(figsize = (14,8))
sns.lineplot(data=flights.pivot(index = "year",columns = "month", values = "passengers"))

plt.xlabel("Año")
plt.ylabel("Pasajeros")
plt.show()
```

Enlace a esta funcion: https://seaborn.pydata.org/generated/seaborn.lineplot.html

### Gráficos de puntos

```python
iris = pd.read_csv("./Datos/iris.csv")
iris.columns = [x.replace(".","_") for x in iris.columns]
```

```python
plt.figure(figsize= (14,8))
sns.scatterplot(data = iris, 
                x = "petal_length", 
                y = "sepal_length", 
                hue = "variety", 
                palette = {"Setosa": "red", "Versicolor": "blue", "Virginica": "green"})
plt.show()
```

Enlace a la documentación de la función: https://seaborn.pydata.org/generated/seaborn.scatterplot.html

### Histogramas 

```python
plt.figure(figsize = (15,8))

sns.histplot(data = iris, x = "petal_width", binwidth=0.1, kde = True)
plt.show()
```

```python
# histograma horizontal
plt.figure(figsize = (15,8))

sns.histplot(data = iris, y = "petal_width", bins = 15)
plt.show()
```

```python
# Histograma por clases
plt.figure(figsize = (15,8))
sns.histplot(data=iris, x="sepal_width", hue="variety", kde = True)
plt.show()
```

Enlace a los histogramas de seaborn: https://seaborn.pydata.org/generated/seaborn.histplot.html

### Graficos de barras

```python
plt.figure(figsize = (15,8))
sns.barplot(data=iris, x="variety", y = "petal_length", estimator = np.median)
plt.show()
```
